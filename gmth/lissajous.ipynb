```python
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# define the time vector (from 0 to 2 seconds, with a step of 0.01 seconds)
t = np.arange(0, 4, 0.001)

# define the parameters (amplitude, frequency, phase) for both sine waves
amp1 = 1.0
freq1 = 2.0
phase1 = 0.0
amp2 = 1.0
freq2 = 3.0
phase2 = 0 * np.pi/2

# calculate the y values for each sine wave
y1 = amp1 * np.sin(2*np.pi*freq1*t + phase1)
y2 = amp2 * np.sin(2*np.pi*freq2*t + phase2)

# Amplitude Modulation (AM) of Sine Wave 1 with Sine Wave 2
y_am = (amp1 + y2) * np.sin(2*np.pi*freq1*t + phase1)

# Frequency Modulation (FM) of Sine Wave 1 with Sine Wave 2
y_fm = amp1 * np.sin(2*np.pi*(freq1 + y2)*t + phase1)

# Create a subplot layout
fig = make_subplots(rows=5, cols=1,
                    subplot_titles=('Lissajous', 'Sine Wave 1', 'Sine Wave 2', 'AM of Sine Wave 1 with Sine Wave 2', 'FM of Sine Wave 1 with Sine Wave 2'))

# Add traces
fig.add_trace(go.Scatter(x=y1, y=y2, mode='lines', line=dict(color='blue')), row=1, col=1)
fig.add_trace(go.Scatter(x=t, y=y1, mode='lines', line=dict(color='black')), row=2, col=1)
fig.add_trace(go.Scatter(x=t, y=y2, mode='lines', line=dict(color='green')), row=3, col=1)
fig.add_trace(go.Scatter(x=t, y=y_am, mode='lines', line=dict(color='red')), row=4, col=1)
fig.add_trace(go.Scatter(x=t, y=y_fm, mode='lines', line=dict(color='purple')), row=5, col=1)

# Update layout
fig.update_layout(height=2000, width=800)

# Update x and y axis titles
titles = [('Sine Wave 1', 'Sine Wave 2'), 
          ('Time (s)', 'Amplitude'),
          ('Time (s)', 'Amplitude'),
          ('Time (s)', 'Amplitude'),
          ('Time (s)', 'Amplitude')]

for idx, (x_title, y_title) in enumerate(titles, start=1):
    fig.update_xaxes(title_text=x_title, row=idx, col=1)
    fig.update_yaxes(title_text=y_title, row=idx, col=1)

# Ensure that the Lissajous figure has a square aspect ratio
fig.update_yaxes(scaleanchor="x", scaleratio=1, row=1, col=1)

# Show plot
fig.show()
```
