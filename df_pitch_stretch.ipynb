```python
import pandas as pd

#load data from TSV

df = pd.read_csv('conga_multidft.tsv', delimiter='\t')

display(df)
```

```python
# Make a copy of the dataframe
df_copy = df.copy()

# Define the stretch factor
stretch_factor = 10

# Define the pitch factor
pitch_factor = 1

# Find the minimum time_start value
min_time_start = df['time_start'].min()

# Multiply the time_start values by the stretch factor, except for the rows with time_start equal to min_time_start
df_copy.loc[df_copy['time_start'] != min_time_start, 'time_start'] *= stretch_factor

# Multiply all time_stop values by the stretch factor
df_copy['time_stop'] *= stretch_factor

# Multiply all freq_start and freq_stop values by the pitch factor
df_copy[['freq_start', 'freq_stop']] *= pitch_factor

display(df_copy)
```

```python
import os
from datetime import datetime

# save the DataFrame as a TSV file
filename = 'conga_stretch.tsv'
try:
    df_copy.to_csv(filename, sep='\t', index=False)
    print(f"Data saved successfully to {os.getcwd()}\\{filename} at {datetime.now()}.")
except:
    print("Error: data could not be saved.")
```
