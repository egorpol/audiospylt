```python
import librosa
from dft_analysis import analyze_signal

# Load audio file
filename = 'bongo.wav'
signal, sr = librosa.load(filename, sr=None, mono=True)

# Analyze signal
peaks_df = analyze_signal(
    signal, 
    sr, 
    filename,
    window_type='hann', 
    thresh_amp_low=0.00016, 
    thresh_amp_high=0.0035, 
    thresh_freq_low=30, 
    thresh_freq_high=4500, 
    prominence=0.0002, 
    width=None,
    show_peaks=True, 
    show_plot=True
)
```

```python
import os
from datetime import datetime

# save the DataFrame as a TSV file
filename = 'conga3.tsv'
try:
    peaks_df.to_csv(filename, sep='\t', index=False)
    print(f"Data saved successfully to {os.getcwd()}\\{filename} at {datetime.now()}.")
except:
    print("Error: data could not be saved.")
```
